```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
import re
```

```python
prices=[]
brands=[]
display_size=[]
Model_ID=[]
Launch_Year=[]
Resolution=[]
Sound=[]

for i in range(2,25):
    print("https://www.flipkart.com/search?q=tv+smart+tv&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=tv+smart+tv%7CTelevisions&requestId=b823f314-e29c-4fb2-9f50-b8fd9dbd88a4&as-searchtext=tv&page="+ str(i))
    urls="https://www.flipkart.com/search?q=tv+smart+tv&sid=ckf%2Cczl&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=RECENT&suggestionId=tv+smart+tv%7CTelevisions&requestId=b823f314-e29c-4fb2-9f50-b8fd9dbd88a4&as-searchtext=tv&page="+str(i)
    page=requests.get(urls)
    soup=BeautifulSoup(page.text)
    
    p=soup.find_all("div",class_="hZ3P6w DeU9vF")
    for i in p:
        prices.append(i.text)
    
    data=soup.find_all("div",class_="RG5Slk")
    for i in data:
        brands.append(i.text.split()[0])
    
    data=soup.find_all("div",class_="RG5Slk")
    for i in data:
        display_size.append(i.text.split("(")[1].split(")")[0])
    
    features=soup.find_all("li",class_="DTBslk")
    for i in features:
        a=re.findall(r"Model ID:\s*(.+)",i.text)
        if len(a)>0:
            Model_ID.append(a[0])
    
    for i in features:
        a=re.findall(r"Launch Year:\s*(\d{4})",i.text)
        if len(a)>0:
            Launch_Year.append(a[0])

    for i in features:
        a=re.findall(r"(Full HD|HD Ready|Ultra HD \(4K\)|Ultra HD \(8K\))",i.text)
        if len(a)>0:
            Resolution.append(a[0])
    
    for i in features:
        a=re.findall(r"Total Sound Output:\s*(\d+\s*W)",i.text)
        if len(a)>0:
            Sound.append(a[0])
    print(len(Resolution))
    print(len(prices))
    print(len(Launch_Year))           
```
